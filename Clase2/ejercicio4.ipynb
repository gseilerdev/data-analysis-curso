```python
# Punto de partida
ventas = [23000, 87000, 45000, 102000, 31500, 68000, 49000, 95000, 12000, 74000]
meta = 50000
total = 0

sobre_meta = []
bajo_meta = []

for venta in ventas:
    total += venta

    if venta < meta:
        bajo_meta.append(venta)
    else:
        sobre_meta.append(venta)

print(sobre_meta)
print(bajo_meta)

total_bajo_meta = sum(bajo_meta)
print(f"Total de ventas bajo la meta: {total_bajo_meta}")

total_sobre_meta = sum(sobre_meta)
print(f"Total de ventas sobre la meta: {total_sobre_meta}")

ventas_totales = total_bajo_meta + total_sobre_meta
print(f"Total de ventas: {ventas_totales}")
```
